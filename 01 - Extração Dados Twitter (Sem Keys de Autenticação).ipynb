```python
from twython import Twython
import json
import pandas as pd
import random
import time
```

```python
#Keys for autentication
consumer_key=''
consumer_secret=''

access_token=''
access_token_secret=''

#Instanciando objeto
python_tweets = Twython(consumer_key, consumer_secret)
```

```python
#Declarando dicionário
dict_ =  {'id':[], 'user': [], 'date' : [], 'text' : [], 'favorite_count': [], 'retweet_count': [], 'location':[]}
```

```python
def buscar_tweets(query_term, max_id, max_iters):
    
    for call in range(0, max_iters):
        #Configurar a query
        query = {
            'q'            : query_term,
            'result_type'      : 'mixed',
            'count'            : 100,
            'lang'             : 'pt',
            'max_id'           : max_id,
            'tweet_mode'       : 'extended',
            'include_entities' : False
        }
        
        for status in python_tweets.search(**query)['statuses']:
            if 'RT @' not in (status['full_text']):
                dict_['id'].append(status['id'])
                dict_['user'].append(status['user']['screen_name'])
                dict_['date'].append(status['created_at'])
                dict_['text'].append(status['full_text'])
                dict_['favorite_count'].append(status['favorite_count'])
                dict_['retweet_count'].append(status['retweet_count'])
                dict_['location'].append(status['user']['location'])
                max_id = status['id']
```

```python
busca = ['windows 11', 'w11', '#windows11']
print('Serão extraídos tweets usando ' + str(len(busca)) + ' termos.')

for item in busca:
    print('Iniciando processamento para ' + item)

    buscar_tweets(item, "", 15)
    print('Final do processamento de {0}. Total de {1} armazenados.'.format(item, len(dict_['id'])))
    randomico   = random.randrange(10, 30)
    print('Vamos aguardar ' + str(randomico) +' segundos.')
    time.sleep(randomico)
    
df = pd.DataFrame(dict_)
df
```

```python
df.to_csv('DadosTwitter.csv', index=False)
```
