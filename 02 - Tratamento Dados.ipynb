```python
import pandas as pd
import os
import glob
from datetime import datetime
```

```python
df  = pd.read_csv('DadosTwitter.csv')
df
```

```python
# Excluindo os registros duplicados
df  = df.drop_duplicates(['text'])
df.count()
```

```python
# Resetando o índice
df = df.reset_index()
df
```

```python
df.drop(columns=['index', 'id', 'user', 'date', 'favorite_count', 'retweet_count', 'location'], inplace=True)
#inplace=True: dataframe assume a versão final após o processamento do comando
```

```python
df
```

```python
#determinando o nome do arquivo
data_hora = datetime.now().strftime('%Y%m%d_%H%M%S')
NomeArquivo = 'TweetsTratados' + data_hora + '.csv'
```

```python
#Salvando o arquivo
df.to_csv(NomeArquivo)
```
